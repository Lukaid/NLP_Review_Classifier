```python
sent = pd.read_table('predict_Coupang_Test_Data_Set_Manual.tsv', encoding='utf-8', header=None)

sent
```

```python
test_data_pred = pd.read_csv('test_topic.csv', encoding='utf-8').iloc[:, 1:]

test_data_pred = pd.concat([test_data_pred, sent[0]], axis = 1)

test_data_pred.columns = ['Review', 'Goods', 'App', 'Delivery', 'Payment', 'CS', 'Price', 'Marketing', 'AD', 'Sentiment']

test_data_pred
```

```python
test_data_pred = test_data_pred[['Review', 'Sentiment', 'Goods', 'App', 'Delivery', 'Payment', 'CS', 'Price', 'Marketing', 'AD']]

test_data_pred
```

```python
print(sum(test_data_pred['Sentiment'] == 'positive'))
print(sum(test_data['Sentiment'] == 'positive'))
```

```python
vaild_list = ['Sentiment', 'Goods', 'App', 'Delivery', 'Payment', 'CS', 'Price', 'Marketing', 'AD']
accuracy = []

for col in vaild_list:
    correct = 0
    for idx in range(len(test_data)):
        if test_data[col][idx] == test_data_pred[col][idx]:
            correct += 1
    accuracy.append(correct / len(test_data))
```

```python
from sklearn.metrics import confusion_matrix
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
vaild_list = ['Sentiment', 'Goods', 'App', 'Delivery', 'Payment', 'CS', 'Price', 'Marketing', 'AD']

for col in vaild_list:
    cm = confusion_matrix(test_data[col], test_data_pred[col])
    print(cm)
    sns.heatmap(cm, annot=True)

sns.heatmap(cm, annot=True)
```

```python
labels = ['True Neg' , 'False Pos' , 'False Neg' , 'True Pos']
labels = np.asarray(labels).reshape(2,2)

sns.heatmap(cm, annot=labels, cmap='Blues', fmt = '')
```

```python
group_names = ['True Neg', 'False Pos', 'False Neg', 'True Pos']
vaild_list = ['Sentiment', 'Goods', 'App', 'Delivery', 'Payment', 'CS', 'Price', 'Marketing', 'AD']
for col in vaild_list:
    cm = confusion_matrix(test_data[col], test_data_pred[col])

    group_counts = ['{0:0.0f}'.format(value) for value in
                    cm.flatten()]
    group_percentages = ['{0:.2%}'.format(value) for value in
                        cm.flatten()/np.sum(cm)]
    labels = [f'{v1}\n{v2}\n{v3}' for v1, v2, v3 in
            zip(group_names,group_counts,group_percentages)]
    labels = np.asarray(labels).reshape(2,2)
    sns.heatmap(cm, annot=labels, fmt='', cmap='Blues').set_title(col)
    plt.savefig(f"{col}_output.png")
    plt.show()
```
